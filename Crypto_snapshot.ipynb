```python

import argparse
import requests
import pandas as pd
from datetime import datetime

API_URL = "https://api.coingecko.com/api/v3/coins/markets"

def fetch_top_coins(vs_currency="usd", per_page=10):
    params = {"vs_currency": vs_currency, "order": "market_cap_desc", "per_page": per_page, "page": 1, "sparkline": "false"}
    r = requests.get(API_URL, params=params, timeout=10)
    r.raise_for_status()
    return r.json()

def save_snapshot(data, out):
    if not data:
        print("[!] No data returned")
        return
    df = pd.DataFrame(data)
    cols = ["id","symbol","name","current_price","market_cap","total_volume","price_change_percentage_24h"]
    selected = [c for c in cols if c in df.columns]
    df[selected].to_csv(out, index=False)
    print(f"[+] Saved {len(df)} rows to {out}")

if __name__ == "__main__":
    parser = argparse.ArgumentParser()
    parser.add_argument("--per-page", type=int, default=10)
    parser.add_argument("--out", default=f"crypto_{datetime.utcnow().strftime('%Y%m%d_%H%M%S')}.csv")

    args, unknown = parser.parse_known_args()
    data = fetch_top_coins(per_page=args.per_page)
    save_snapshot(data, args.out)
```
